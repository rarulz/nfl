```python
#Mapping Top 3 team valuations
import matplotlib.pyplot as plt

years = [2016,2017,2018,2019,2020]
arr1  = [4,4.2,4.8,5,5.5]
arr2  = [3.65,	3.58,	4.09,	4.24,	4.24]
arr3  = [3.55,	3.64,	4.06,	4.02,	4.02]

# Stack Plot Legend.
plt.plot([], [], color='r', label = 'Dallas Cowboys')
plt.plot([], [], color='g', label = 'Real Madrid')
plt.plot([], [], color='b', label = 'FC Barcelona')

plt.stackplot(years, arr1, arr2, arr3, colors= ['r', 'g', 'b'])
plt.title('Top 3 Team Combined Valuations')
plt.xlabel('Years')
plt.ylabel('Valuation in USD Billions')
plt.legend()
plt.show()
```
